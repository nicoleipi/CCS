```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Read Data
data_900_raw = pd.read_csv('model1_csv/model1_data_900_transposed.csv')
data_900 = data_900_raw.round(2)
data_900
```

```python
#Data
# one = 6-31G(d), two = 6-31G(dp), three = 6-31+G(dp), four = 6-311G(dp), five = 6-311G(2df2pd)
# six = 6-311++G(dp), seven = 6-311++G(2df2pd)
energy = data_900.energy
one = data_900.one
two = data_900.two
three = data_900.three
four = data_900.four
five = data_900.five
six = data_900.six
seven = data_900.seven

#Aesthetics
fig = plt.figure()
fig.patch.set_facecolor('white')

#Plotting
font = {'fontname':'DejaVu Sans'}
w=0.1

bar1 = np.arange(len(energy))
plt.barh(bar1, one-seven, w, color='coral', align='center', label='6-31G(d)')

bar2 = bar2 = [i-w for i in bar1]
plt.barh(bar2, two-seven, w, color='lightblue', align='center', label='6-31G(dp)')

bar3 = bar3 = [i-2*w for i in bar1]
plt.barh(bar3, three-seven, w, color='grey', align='center', label='6-31+G(dp)')

bar4 = bar4 = [i-3*w for i in bar1]
plt.barh(bar4, four-seven, w, color='tan', align='center', label='6-311G(dp)')

bar5 = bar5 = [i-4*w for i in bar1]
plt.barh(bar5, six-seven, w, color='teal', align='center', label='6-311++G(dp)')

bar6 = bar6 = [i-5*w for i in bar1]
plt.barh(bar6, five-seven, w, color='sienna', align='center', label='6-311G(2df2pd)')

bar7 = bar7 = [i-6*w for i in bar1]
plt.barh(bar7, seven-seven, w, color='aquamarine', align='center') #, label='6-311++G(2df2pd)')

plt.legend(bbox_to_anchor=(0.37, 1, 1, 0), loc=1, ncol=1, borderaxespad=0)

#Value Labels
def addlabels(x,y,v):
    for i in range(len(x)):
        plt.text(y[i]-0.098, i-v,y[i], ha="right", va="bottom")
        
#Axes        
ax = plt.gca()
ax.set_facecolor('#ffffff')
ax.set_xlim([0,0.32])
plt.title('Energy Calculations (Hartree) Compared to Basis Set for Model 1 Isomer 900', font)
plt.ylabel('Energy Calculation', font)
plt.xlabel('Energy (Hartree) with 6-311++G(2df2pd) as Reference', font)
plt.yticks(np.arange(5), ['E+ZPE', 'U', 'H', 'G', 'E'])

#Show
plt.savefig('model1_900_hartree.png', bbox_inches='tight', dpi=300)
plt.show()
```
