```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import csv
import math
```

B3LYP M1

```python
# declare variables
num_files = 16
num_rows = 7
num_columns = 2
#kcal_conv = 627.5095 #kcal/mol per 1 hartree
rt = 298.15*(0.001987/627.5095) #kcal/mol

# declare arrays
raw_data = np.zeros((num_files, num_rows, num_columns))
#print(raw_data)
input_filename = []

directory = r'data/model1/B3LYP/isomer'
file_id = 0
for file in os.listdir(directory):
    full_path = os.path.abspath(directory)
    real_file = full_path + "\\" + file
    input_filename.append(file)

    results = []

    with open(real_file) as csvfile:
        reader = csv.reader(csvfile, delimiter=',', quotechar='|') 
        row_id = 0
        for row in reader:
            if (row_id > 0):
                raw_data [file_id][row_id-1][0] = row[4]
                raw_data [file_id][row_id-1][1] = row[5]
            row_id = row_id + 1
    file_id = file_id + 1
#print(input_filename)
#print(raw_data)

def eq_frac(input_file, input_row, input_column):
    #print(-raw_data[input_file][input_row][input_column]/rt)
    numerator = -raw_data[input_file][input_row][input_column]/rt
    denominator = 0
    for i in range (0, 16):
        denominator = denominator + np.exp((-raw_data[i][input_row][input_column]/rt)-numerator)
    return(1.0/denominator)

raw_output = np.zeros((num_files, num_rows, num_columns))
for file_id in range (0, 16):
    for row_id in range (0, 7):
        raw_output [file_id][row_id][0] = eq_frac(file_id, row_id, 0)
        raw_output [file_id][row_id][1] = eq_frac(file_id, row_id, 1)

for file_id in range (0,16):
    print(input_filename[file_id])
    print(raw_output[file_id])
    print("")
    pd.DataFrame(raw_output[file_id]).to_csv('B3LYP_model1/calc/%s' % input_filename[file_id])
```
