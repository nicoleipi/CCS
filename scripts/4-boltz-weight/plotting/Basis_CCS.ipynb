```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import os  
```

```python
#Aesthetics
mpl.rcParams['axes.prop_cycle'] = mpl.cycler(color=["c", "cornflowerblue", "goldenrod", "darkorange"]) 
fig = plt.figure(figsize=(13, 6))
fig.patch.set_facecolor('white')

#Plotting
font = {'fontname':'DejaVu Sans',}

#Read Data
data = pd.read_csv('data/ccs/B3LYP_model1.csv')
data = data.round(2)

#Data
G = data.CCS_G
E = data.CCS_E

plt.plot(data.index.values*100, G, label='B3LYP G', marker='o')
plt.plot(data.index.values*100, E, label='B3LYP E', marker='o')

#Read Data
data = pd.read_csv('data/ccs/D3BJ_model1.csv')
data = data.round(2)

#Data
G = data.CCS_G
E = data.CCS_E

plt.plot(data.index.values*100, G, label='B3LYP-D3(BJ) G', marker='o')
plt.plot(data.index.values*100, E, label='B3LYP-D3(BJ) E', marker='o')

plt.legend(loc = 'best', ncol = 5)
 
#Axes
ax = plt.gca()
ax.set_facecolor('#ffffff')
ax.set_ylim([132,138])
plt.title('Model 1 Boltzman Weighted Collision Cross Section', font)
plt.xlabel('Basis Set', font, fontweight='bold')
plt.ylabel('Boltzmann Weighted CCS ($\mathregular{m^{2}}$)', font, fontweight='bold')
plt.xticks([0, 100, 200, 300, 400, 500, 600], ['6-31G(d)', '6-31G(d,p)', '6-311G(d,p)', '6-311G(2df,2pd)', '6-31+G(d,p)', '6-311++G(d,p)', '6-311++G(2df,2pd)'])

#Show
plt.savefig('plots/ccs/model1.png', bbox_inches='tight', dpi=300)
plt.show()
```
