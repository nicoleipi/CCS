```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Read Data
data_683_raw = pd.read_csv('model1_data_683.csv')
data_683 = data_683_raw.round(2)

#Data
basis_set = data_683.Basis
EZPE = data_683.EZPE
U = data_683.U
H = data_683.H
G = data_683.G
E = data_683.E

#Plotting
font = {'fontname':'DejaVu Sans'}
w=0.1

bar1 = np.arange(len(basis_set))
plt.barh(bar1, EZPE, w, color='#f94144ff', align='center', label='E+ZPE')

bar2 = bar2 = [i-w for i in bar1]
plt.barh(bar2, U, w, color='#45cb34ff', align='center', label='U')

bar3 = bar3 = [i-2*w for i in bar1]
plt.barh(bar3, H, w, color='#ffca4cff', align='center', label='H')

bar4 = bar4 = [i-3*w for i in bar1]
plt.barh(bar4, G, w, color='#28b4f1ff', align='center', label='G')

bar5 = bar5 = [i-4*w for i in bar1]
plt.barh(bar5, E, w, color='#f8961eff', align='center', label='E')

plt.legend(prop={'size': 7.5})

#Value Labels
def addlabels(x,y,v):
    for i in range(len(x)):
        plt.text(y[i]-0.098, i-v,y[i], ha="right", va="bottom")
        
#addlabels(bar1, EZPE, 0.1)
#addlabels(bar2, U, 0.1)
#addlabels(bar3, H, 0.1)
#addlabels(bar4, G, 0.1)
#addlabels(bar5, E, 0.1)
        
#Axes        
axes = plt.gca()
axes.set_xlim([-795.5,-796.5])
plt.title('Energy Calculations Compared to Basis Set for Model 1 Isomer 683', font)
plt.ylabel('Basis Set', font)
plt.yticks(np.arange(7), ['6-31G(d)', '6-31+G(dp)', '6-311G(2df2pd)', '6-31G(dp)', '6-311++G(2df2pd)', '6-311G(dp)', '6-311++G(dp)'])

#Show
plt.savefig('model1_683', bbox_inches='tight', dpi=300)
plt.show()
```
