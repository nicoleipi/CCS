```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
directory = r'D3BJ_model4/csv/'
for file in os.listdir(directory):
    #print(file)
    filename = os.path.splitext(file)[0]
    #print(filename)

    isomer = filename.split("_",4)
    #print(isomer[3])
    
    #Read Data
    data_raw = pd.read_csv('D3BJ_model4/csv/%s' % file)
    data = data_raw.round(2)
    #print(data)

    #Data
    basis_set = data.Basis
    EZPE = data.EZPE
    U = data.U
    H = data.H
    G = data.G
    E = data.E

    #Aesthetics
    fig = plt.figure()
    fig.patch.set_facecolor('white')

    #Plotting
    font = {'fontname':'DejaVu Sans'}
    w=0.1

    bar1 = np.arange(len(basis_set))
    plt.barh(bar1, EZPE, w, color='#f94144ff', align='center', label='E+ZPE')

    bar2 = bar2 = [i-w for i in bar1]
    plt.barh(bar2, U, w, color='#45cb34ff', align='center', label='U')

    bar3 = bar3 = [i-2*w for i in bar1]
    plt.barh(bar3, H, w, color='#ffca4cff', align='center', label='H')

    bar4 = bar4 = [i-3*w for i in bar1]
    plt.barh(bar4, G, w, color='#28b4f1ff', align='center', label='G')

    bar5 = bar5 = [i-4*w for i in bar1]
    plt.barh(bar5, E, w, color='#f8961eff', align='center', label='E')

    plt.legend(prop={'size': 7.5})

    #Value Labels
    def addlabels(x,y,v):
        for i in range(len(x)):
            plt.text(y[i]-0.098, i-v,y[i], ha="right", va="bottom")
            
    #addlabels(bar1, EZPE, 0.1)
    #addlabels(bar2, U, 0.1)
    #addlabels(bar3, H, 0.1)
    #addlabels(bar4, G, 0.1)
    #addlabels(bar5, E, 0.1)
            
    #Axes        
    ax = plt.gca()
    ax.set_facecolor('#ffffff')
    ax.set_xlim([-795.8,-796.9])
    plt.title('Energy Calculations (Hartree) Using B3LYP-D3(BJ) Compared to Basis Set for Model 4 Isomer %s' % isomer[3], font)
    plt.ylabel('Basis Set', font)
    plt.xlabel('Hartree', font)
    plt.yticks(np.arange(7), ['6-31G(d)', '6-31G(d,p)', '6-31+G(d,p)', '6-311G(d,p)', '6-311++G(d,p)', '6-311G(2df,2pd)', '6-311++G(2df,2pd)'])

    #Show
    plt.savefig('B3LYP_model4/plots/B3LYP_model4_%s.png' % isomer[3], bbox_inches='tight', dpi=300)
    plt.show()
```
