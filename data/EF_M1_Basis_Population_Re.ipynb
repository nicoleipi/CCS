```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import os  
```

B3LYP Gibb's Free Energy

```python
#Aesthetics
mpl.rcParams['axes.prop_cycle'] = mpl.cycler(color=["c", "seagreen", "cornflowerblue", "darkorange", "goldenrod"]) 
fig = plt.figure(figsize=(13, 6))
fig.patch.set_facecolor('white')

#Plotting
font = {'fontname':'DejaVu Sans',}

directory = r'B3LYP_model1/reformatted_calc'
for file in os.listdir(directory):
    #print(file)
    filename = os.path.splitext(file)[0]
    #print(filename)

    isomer = filename.split("_",4)
    
    #Read Data
    data_raw = pd.read_csv('B3LYP_model1/reformatted_calc/%s' % file)
    xdata = data_raw*100
    data = xdata.round(2)
    #print(data)
    
    #Data
    basis = data.basis_set
    G = data.G
    E = data.E

    plt.plot(basis, G, label=isomer[3], marker='o')

plt.legend(loc = 'best', ncol = 5)
 
#Axes
ax = plt.gca()
ax.set_facecolor('#ffffff')
ax.set_ylim([0,100])
plt.title('Model 1 Relative Populations (B3LYP) Gibb\'s Free Energy', font)
plt.xlabel('Basis Set', font, fontweight='bold')
plt.ylabel('Probablility (%)', font, fontweight='bold')
plt.xticks([0, 100, 200, 300, 400, 500, 600], ['6-31G(d)', '6-31G(d,p)', '6-311G(d,p)', '6-311G(2df,2pd)', '6-31+G(d,p)', '6-311++G(d,p)', '6-311++G(2df,2pd)'])

#Show
plt.savefig('model1/b_plots/g_re_model1.png' % basis[1], bbox_inches='tight', dpi=300)
plt.show()
```
