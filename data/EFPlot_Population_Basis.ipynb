```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

B3LYP M1

```python
directory = r'B3LYP_model1/calc/'
for file in os.listdir(directory):
    #print(file)
    filename = os.path.splitext(file)[0]
    #print(filename)

    isomer = filename.split("_",4)
    #print(isomer[3])
    
    #Read Data
    data_raw = pd.read_csv('B3LYP_model1/calc/%s' % file)
    xdata = data_raw*100
    data = xdata.round(2)
    #print(data)
    
    #Data
    basis_set = data.basis_set
    G = data.G
    #correct_G = G.round(3)
    E = data.E
    #correct_E = G.round(3)

    #Aesthetics
    fig = plt.figure()
    fig.patch.set_facecolor('white')

    #Plotting
    font = {'fontname':'DejaVu Sans',}
    w=0.3

    bar1 = np.arange(len(basis_set))
    plt.barh(bar1, G, w, color='#28b4f1ff', align='center', label='G')

    bar2 = bar2 = [i-w for i in bar1]
    plt.barh(bar2, E, w, color='#f8961eff', align='center', label='E')

    plt.legend(prop={'size': 10})

    #Value Labels
    def addlabels(x,y,v):
        for i in range(len(x)):
            plt.text(y[i]+7.5, i-v,y[i], ha="right", va="bottom", fontsize=8)
            
    addlabels(bar1, G, 0.15)
    addlabels(bar2, E, 0.45)
            
    #Axes        
    ax = plt.gca()
    ax.set_facecolor('#ffffff')
    ax.set_xlim([0,100])
    plt.title('Equilbriam Fraction of Model 1 Isomer %s (B3LYP)' % isomer[3], font)
    plt.ylabel('Basis Set', font)
    plt.xlabel('Probablility (%)', font)
    plt.yticks(np.arange(7), ['6-31G(d)', '6-31G(d,p)', '6-31+G(d,p)', '6-311G(d,p)', '6-311++G(d,p)', '6-311G(2df,2pd)', '6-311++G(2df,2pd)'])

    #Show
    plt.savefig('B3LYP_model1/calc_plots/B3LYP_model1_%s.png' % isomer[3], bbox_inches='tight', dpi=300)
    plt.show()
```
